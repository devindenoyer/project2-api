````python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin

url = "https://en.wikipedia.org/wiki/The_Cure"  
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
img_tag = soup.find('img', {'alt':"The Cure performing in August 2007. From left to right: Jason Cooper, Porl Thompson, Robert Smith, and Simon Gallup"})
img_url = img_tag['src']
if not img_url.startswith(('http://', 'https://')):
    img_url = urljoin(url, img_url)
img_response = requests.get(img_url)
img_path = os.path.join(os.getcwd(), 'downloaded_image.jpg')
with open(img_path, 'wb') as file:
    file.write(img_response.content)

print(f"Image downloaded and saved to {img_path}")
````

````python

def clean(text):
    text = text.lower()
    text = ''.join([char for char in text if char.isalpha() or char.isspace()])
    text = ' '.join(text.split())
    return text

clean(soup.text)
````

````python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags
def fetch_text_content(url):
    try:
        parsed_url = urlparse(url)
        if not parsed_url.scheme:
            url = urljoin('http://', url)
        if not parsed_url.netloc:
            print(f"Invalid URL '{url}': No host supplied")
            return ""

        response = requests.get(url)
        response.raise_for_status()  

        if 'text/html' not in response.headers.get('Content-Type', ''):
            print(f"Skipping non-HTML content at {url}")
            return ""

        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        text_content = soup.get_text()

        return text_content[:1000]
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")
        return ""

scripts['text_content'] = scripts['link'].apply(fetch_text_content)
scripts.to_csv('pudding_texts_cleaned.csv', index=False)
print(scripts)
````

*For this code, I utalized some technqiues I used previously in part 2, but I also utalized ChatGPT which allowed me to fetch the JSON reponse from the specificed API endpoint and extract data from this response. 

````python

url = "https://api.themoviedb.org/3/movie/now_playing"  
api_key = "83f3f3d617cd11851c1707cc5ddd9e87"  

response = requests.get(url, params={"api_key": api_key})
json_response = response.json()
print(json_response)

movies = json_response.get('results', [])

df = pd.DataFrame(movies)
metadata_fields = ['title', 'release_date', 'overview', 'vote_average', 'popularity', 'original_language', 'poster_path']
df = df[metadata_fields]

print(df)

def download_poster(poster_path, title):
    base_url = "https://image.tmdb.org/t/p/w500" 
    poster_url = f"{base_url}{poster_path}"
    response = requests.get(poster_url)
    if response.status_code == 200:
        valid_title = "".join(c for c in title if c.isalnum() or c in (' ', '_')).rstrip()
        file_path = os.path.join(os.getcwd(), f"{valid_title}.jpg")
        with open(file_path, 'wb') as file:
            file.write(response.content)
        print(f"Poster for '{title}' downloaded and saved to {file_path}")
    else:
        print(f"Failed to download poster for '{title}'")
for i, row in df.head(10).iterrows():
    poster_path = row['poster_path']
    title = row['title']
    if poster_path:
        download_poster(poster_path, title)
    else:
        print(f"No poster available for '{title}'")
````
